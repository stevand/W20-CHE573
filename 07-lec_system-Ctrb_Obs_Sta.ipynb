## Discrete System Properties



In the previous sections, we discussed $z$-transform. In this section, we discuss important properties in the analysis of linear time invariant systems. In particular we will address three important systems properties **stability**, **controllability** and **observability**.



### Stability



-   **Question**: *What is the condition for stability of the discrete linear system*?
    This question is one of the most important questions which needs  to be answered for the understanding of the qualitative system behaviour. Stability is  a fundamental system theoretic property. In general, there are several useful criteria to determine stability of the linear time invariant discrete/continuous system. We will concentrate on the following ones:
    -   Direct numerical evaluation of the eigenvalues of the matrix $\Phi$. The matrix $\Phi$ represents the autonomous system dynamics (that is $u(z)=0$) and determines the response of the system without any  input being applied. Therefore, we have:
        $$ x(k+1)=\Phi x(k),\;\;x(0)=x_0\rightarrow \;\;x(k)=\Phi^kx_0$$
        
        ![img](./FiguresCHE573/stability_radius1.png "Stability radius")
        
        Stability regions in $s$ and $z$-plane. A simple relation among continuous and discrete system representations is given as $z=e^{s\Delta t}$. Since $s$ is a complex variable, one can write $s=\alpha+j\beta$, which implies that $z=e^{\alpha\Delta t}e^{j\beta\Delta t}=e^{\alpha\Delta t}(cos(\beta \Delta t)+jsin(\beta \Delta t))$. Thus, for real values of $s$, the magnitude of $z$ is $|z|=e^{\alpha\Delta t}$, and implies that $e^{\alpha\Delta t}>1$ for $\alpha>0$, $e^{\alpha\Delta t}=1$ for $\alpha=0$ and $e^{\alpha\Delta t}<1$ for $\alpha<0$
        
        However, we do not know what is the expression for $\Phi^k$ (in the same way for continuous system we have that $x(t)=e^{At}x(0)=\left[?\right]x(0)$). Therefore, we can find the state transformation $z(k)=Tx(k)$, so that $D=T\Phi T^{-1}$ is such that
        above Equation becomes:
        $$z(k)=\left[T \Phi T^{-1}\right]z(0) \;\;\;\; \rightarrow z(k)=D^kz(0)=\left[\begin{matrix}\lambda^k_1 & 0 & \cdots& 0\\0 & \lambda^k_2 & 0 & \cdots\\\cdots & \cdots& \cdots &\cdots\\0 & 0& 0& \lambda^k_n \end{matrix}\right]z(0)$$
        
        In order to determine the system's stability (if the state $x(k)$ diverges  or converges for $x_0\neq 0$, as $k\rightarrow \infty$), we have the following criteria: *If* $\Phi$ *is diagonalizable, the necessary condition for above Eq. stability is that all eigenvalues of* $\Phi$, $\|\left\{\lambda_i\right\}\|<1$.
        
        The eigenvalues of $\Phi$ are calculated numerically from  the characteristic polynomial of $\Phi$, that is $\Pi(\lambda)=det(A-\lambda I))=0$,
        $$ \Pi(\lambda)=\lambda^n-p_1\lambda^{n-1}+\cdots+p_{n-1}\lambda +p_n$$
        
        The above Eq. is stable if and only if $\|\lambda\|<1$ for every characteristic root, (that is, for every zero of the characteristic polynomial, see [http://www.youtube.com/watch?v=i9c8MG2KW-g](http://www.youtube.com/watch?v=i9c8MG2KW-g).

Determine stability of $x(k+1)=Ax(k)$, where
    $$A=\left[\begin{matrix} 1 & 1\\0.25 &1\end{matrix}\right]$$.

Easily we find

-   Algebraic criteria relate to the properties of the characteristic polynomial (Yuri criteria). Given the characteristic equation in the following form:
    $$\Pi(\lambda)=a_n\lambda^n+a_{n-1}\lambda^{n-1}+\cdots+a_{1}\lambda+a_0$$

The Yuri stability test requires that the following conditions are satisfied in order to guarantee stability:

-   $\Pi(1)>0$
-   $(-1)^n\Pi(-1)>0$ for $n$ even
-   $(-1)^n\Pi(-1)<0$ for $n$ odd
-   and $|a_0|<a_n$, $|b_0|>|b_{n-1}|$, $|c_0|>|c_{n-2}|$,&#x2026;,$|s_0|>|s_{2}|$

  A quick method to determine the stability of the second order system, $\Pi(\lambda)=a_2\lambda^2+a_1 \lambda+a_0$,
is that the following holds:

-   $\Pi(1)>0$
-   $\Pi(-1)>0$
-   $|a_0|<a_2$.
    **Remark**: The alternative technique to the Yury's test that deals with the stability of the characteristic equation is a *modified Routh stability test* which is the stability test to analyze a characteristic equation of the continuous time system. Given the characteristic equation $\Pi(z)=a_nz^n+a_{n-1}z^{n-1}+\cdots+a_{1}z+a_0$ from $\Pi(z)=det(zI-\Phi)$, one can substitute for $z=\frac{1+w}{1-w}$ and obtain the corresponding $\Pi(w)$ which then can be used by standard Routh array test which requires that for stability all coefficients in the Routh array must be positive.

In $Z$-domain, a relaxed LTI system is BIBO stable if and only if its poles of the transfer function $H(z)$ are inside the unit circle. This is a very efficient approach to check the stability of an LTI system. The result is illustrated in the following graph.

![img](./FiguresCHE573/stability_radius2.png "Stability radius")

In CHE 446 Process Control, we learned that for continuous-time systems, if the poles of the Laplace transfer function of a system are on the left half plane of the complex $s$-plane, the system is stable. The difference between the results for continuous-time system and discrete-time systems actually can be explained by studying the relation between Laplace transform and $z$-transform.

Let us recall the definition for Laplace transform. The Laplace transform of a continuous-time signal $x_c(t)$ is defined as follows:

$$X_c(s) = \int_{t=0}^{t=\infty}x_c(t)e^{-st}dt$$

The $z$-transform of a discrete-time signal $x(n)$ is defined as follows:

$$X(z) = \sum_{n=0}^{\infty}x(n)z^{-n}$$

Let us also consider that $x(n)$ is the sampled signal of $x_c(t)$ with the sampling time $T$. Then,

$$x(n)=x_c(nT)$$

Substituting this discretization time back to, we can have:

$$X(z) = \sum_{n=0}^{\infty}x_c(nT)z^{-n}$$

By comparing above equations, it can be seen that:

$$X(z) = X_c(s)|_{t=nT}$$

with

$$z=e^{sT}$$

we can interpret $z$-transform as the sampled Laplace transform with $z=e^{sT}$.

Let us consider an area on the left half of the $s$-plane as shown in the following graph. The area is with the boundary $A\rightarrow B\rightarrow C\rightarrow C'\rightarrow D\rightarrow E'\rightarrow E\rightarrow F\rightarrow A$. In the graph, $w_s=\dfrac{2\pi}{T}$ where $T$ is the sampling time. The points on the boundary are mapped to the $z$-plane as follows:

-   Point A: $s=0$ $\rightarrow$ $z=e^{T\cdot 0}=1$
-   Point B: $s=i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=e^{T\cdot s}=\cos(\dfrac{Tw_s}{2})+i\sin(\dfrac{Tw_s}{2})=\cos(\pi)+i\sin(\pi)=-1$
-   Point C: $s=-\alpha+i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=-e^{-T\alpha}$
-   Point C': $s=-\alpha+i\cdot \dfrac{w_s}{4}$ $\rightarrow$ $z=e^{-T\alpha}(\cos(\dfrac{\pi}{2})+i\sin(\dfrac{\pi}{2}))=i\cdot e^{-T\alpha}$
-   Point D: $s=-\alpha$ $\rightarrow$ $z=e^{-T\alpha}$
-   Point E': $s=-\alpha-i\cdot \dfrac{w_s}{4}$ $\rightarrow$ $z=-i\cdot e^{-T\alpha}$
-   Point E: $s=-\alpha-i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=-e^{-T\alpha}$
-   Point F: $s=-i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=-1$

It can be seen that the area in the $s$-plane is mapped into an annulus with outside radius $r_o=1$ and inside radius $r_i=e^{-T\alpha}$. Similar analysis can be applied to other areas. What we should find is that the left half of the $s$-plane is mapped to the inside of the unit circle. This explains why in the discrete-time domain, the stability condition is poles are within the unit circle.

## Controllability



-   The question to be answered when it comse to the controlalbility of the system is: *Can we find the input function* *(u(t)\\) /such that the system is transfered from initial state* $x(0)$ *to arbitrary desired state in finite time?*
    -   In the oil industry large interconnections of the tank and vessel systems are used for the storage and inventory management. We are interested if we can use the input $u(k)$ to drive oil to the tanker. The answer is given by answering whether the system (tanks and vessels) is controllable from the input $u(k)$, or in other words, is there any input that can drive the system state to a desired point in the finite time. Practically, we want to provide an input $u(k)$ such that the tank  height  (T9) is at a certain level which will provide the pressure  gradient  based force to fill the tanker (see the figure in the notes).
    
    -   **Controllability** is an important system's theoretic property. The controllability in a layman's language answers as to whether the system can be controlled. Let us look at the discrete state space description, given as:
        
        $${{x}(k+1)}=A x(k)+B u(k)\\y(k)=Cx(k)+Du(k)$$
        
        the system is controllable if $x(0)$ can be transferred to a final state $x^{*}$ in the finite time $t=nT$ regardless of $x(0)$. Therefore, if the system is controllable, there exists some $u(t)=u(nT)=u(n)$ such that the system will be transferred from some arbitrary initial state by the input $u(k)$ to some final state in the finite time. The controllability is usualy defined (to the origin) and reachability (from the origin). Therefore, we want to transfer $x(0)\rightarrow x^*=x(N)$ by the application of the set of inputs given as:
        $$\left\{u(0),u(1),u(2),u(3),\cdots,u(k),\cdots,u(N-1)\right\}$$
        
        Hence, we are interested in the existence of the $u(k)$ that will transfer the system and we provide a definition to check controllability (Pontrygin).
        
        **Controllability (Def. by Pontrygin**: *\`\`The system is controllable if and only if the rank of matrix*
        $$rank\left[ B\; AB\;A^2B\;\cdots A^{N-1}B\right]=n$$
        *where $n$ is the size of matrix A*
        
        Let us construct the sequence of control moves
        $$\left\{u(0),u(1),u(2),u(3),\cdots,u(k),\cdots,u(N-1)\right\}$$
        which transfer the system from $x(0)\rightarrow x(N)=x^*$,
        
        $$\begin{array}x(0)&&\nonumber\\x(1)&=&Ax(0)+Bu(0)\nonumber\\x(2)&=&A^2x(0)+ABu(0)+Bu(1)\nonumber\\x(3)&=&A^3x(0)+A^2Bu(0)+ABu(1)+Bu(2)\nonumber\\\cdots&=&\cdots\nonumber\\x^*=x(N)&=&A^Nx(0)+A^{N-1}Bu(0)+A^{N-2}Bu(1)+\cdots+ABu(N-2)+Bu(N-1)\end{array}$$

  Now, let us rewrite the last expression such that the initial condition contribution is on the left side of the equation, therefore we
get

$$x^*-A^Nx(0)=\left[B\;AB\;A^2B\cdots A^{N-1}B\right]\left[\begin{array}{c}u(N-1)\\u(N-2)\\u(N-3)\\\cdots\\u(1)\\u(0)\end{array}\right]$$

Given the vector $x^*$, $A$ and $x(0)$, these are two vectors in $R^n$, and  $A^n\in n \times n$, so the result of $\bar{x}=x^*-A^Nx(0)$ must be in the range of the matrix $C_{trb}$. We know that the system of equations $C_{trb}\textbf{u}=\bar{x}$ has a solution only if $C_{trb}$ is invertible, which implies that the columns of $C_{trb}$ must have $n$-independent vectors. Therefore,
$$rank\left[B\;AB\;A^2B\cdots A^{n-1}B\right]=n$$

The range of matrix
$$C_{trb}=\left[B\;AB\;A^2B\cdots A^{n-1}B\right] $$
is called *controllable subspace*. Naturally, from the above equation we observe that small increments of state trajectories must be independent, otherwise the path is the function of direction, see [http://www.youtube.com/watch?v=KvTm6qRtffM](http://www.youtube.com/watch?v=KvTm6qRtffM).

Finally, we list the following three useful complementary criteria to check controllability of the system.

Controllability conditions}}: The system is controllable if one of these three equivalent conditions holds:

-   rank $[B \;AB\;A^2B\cdots A^{n-1}B]=n$ where $n$ is the size of $A$-matrix
-   rank $[\lambda I-A |B]=n$, $\forall$ $\lambda$ eigenvalue of matrix $A$,  $$lambda  \in C$-complex plane
-   rank $W_n=n$

We notice in the second condition when the $\lambda$ is an eigenvalue of the matrix $A$ the rank will be less than $n$.
Also, one can ask oneself  *How much is the system controllable?* - The answer to that question is coming from the properties of the controllability Gramian, since if $W_n$ is close to drop the rank, then the inverse is large and therefore it requires high efforts to control  the system.

It is important to remember that controllability guarantees the ability to transfer  the  initial state to the final one by some control input action over a finite time.  In addition to controllability, another notion used here is *reachability*. Reachability is defined as an ability of the system to be transferred from the origin to a desired state, while controllability is as an ability to transfer system from an arbitrary state to the origin. Reachability implies controllability, while controllability implies reachability when the state transition matrix $A^n$ is nonsingular.



### Determining controllability



Determine controllability of the following discrete system:
   $$\begin{array}{ccc} x_1(k+1)&=&ax_1(k)+bx_2(k)\\ x_2(k+1)&=&cx_1(k)+dx_2(k)+u(k)\end{array}$$
   with $ab-cd\neq 0$. Therefore, we have
   $$A=\left[\begin{matrix}a& b\\c &d\end{matrix}\right]$$
   and
   $$B=\left[\begin{matrix}0\\1\end{matrix}\right]$$
   so that the condition for controllability is:
   $$ C_{ctrb}=\left[\begin{matrix}B&AB\end{matrix}\right]=\left[\begin{matrix}0&b\\1&d\end{matrix}\right]$$
   which has rank 2 only iff $b\neq 0$
   Another way to check controllability for the discrete systems is to check *controllability Gramian (Wronskian)*, which is given as:
   $$W_n=\sum_{i=0}^n A^iBB'A'^i$$
   so that the following claim is made: *The system is controllable if and only if rank* $W_n=n$



### Determining controllability example



For the two-tank system find the $u(k)$ such that the initial value of the tank heights $h_1=0$ and $h_2=0$ will reach $h_1=1.5$ and $h_2=5$, where $A_1=2$, $A_2=1$, $K_1=K_2=1$ and $\rho=1$, and the sampling time $T=1$.

The state space realization or modelling of the two-tank system lead to the following set of equations:

$$\frac{d}{dt}\begin{array}{ccc}\left[ \begin{matrix} h_1(t)\\ h_2(t)\end{matrix}\right]&=&\left[\begin{matrix} -\frac{K_1}{\rho A_1}&0 \\ \frac{K_1}{\rho A_2}&-\frac{K_2}{\rho A_2}\end{matrix}\right]\left[ \begin{matrix} h_1(t)\\ h_2(t)\end{matrix}\right]+\left[ \begin{matrix} \frac{1}{\rho A_1}\\ 0\end{matrix}\right]u(t)\end{array}$$

We obtain discrete representation which leads to:

$$\begin{array}{ccc}\left[ \begin{matrix} h_1(k+1)\\ h_2(k+1)\end{matrix}\right] &=&\left[\begin{matrix} 0.6065 & 0 \\ 0.4773 & 0.3679\end{matrix}\right] \left[\begin{matrix} h_1(k)\\ h_2(k)\end{matrix}\right]+\left[ \begin{matrix} 0.3935\\ 0\end{matrix}\right]u(k)\end{array}$$

Calculation of the controllability matrix leads to the following expression:

$$\begin{array}{ccc} C_{trb}&=& \left[ \begin{matrix} 0.3935 & 0.2387\\ 0 & 0.1878\end{matrix}\right]  \end{array}$$

therefore $rank(C_{trb})=2$, and the system is controllable, so that  we can find a $u(k)$ as:

$$\begin{array}{ccc} x_0&=&x(0)\nonumber\\ x(1)&=&Ax(0)+Bu(0)\nonumber\\ x(2)&=&Ax(1)+[B\;AB][u(0)\;\;u(1)]'\nonumber \end{array}$$

which leads to

$$[u(0)\;\;u(1)]' = [B\;AB]^{-1}(x(2)-A^2x(0))=[-12.33\;\;26.62]'$$

**Remark**: One should notice that the heights in both tanks become very large at first instance $x(1)$ and surprisingly for this system, the path of the state evolution does not take an intuitively straight line from the origin to the final state.

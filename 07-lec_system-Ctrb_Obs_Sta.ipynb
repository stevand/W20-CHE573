## Discrete System Properties



In the previous sections, we discussed $z$-transform. In this section, we discuss important properties in the analysis of linear time invariant systems. In particular we will address three important systems properties **stability**, **controllability** and **observability**.



### Stability



-   **Question**: *What is the condition for stability of the discrete linear system*?
    This question is one of the most important questions which needs  to be answered for the understanding of the qualitative system behaviour. Stability is  a fundamental system theoretic property. In general, there are several useful criteria to determine stability of the linear time invariant discrete/continuous system. We will concentrate on the following ones:
    -   Direct numerical evaluation of the eigenvalues of the matrix $\Phi$. The matrix $\Phi$ represents the autonomous system dynamics (that is $u(z)=0$) and determines the response of the system without any  input being applied. Therefore, we have:
        $$ x(k+1)=\Phi x(k),\;\;x(0)=x_0\rightarrow \;\;x(k)=\Phi^kx_0$$
        
        ![img](./FiguresCHE573/stability_radius1.png "Stability radius")
        
        Stability regions in $s$ and $z$-plane. A simple relation among continuous and discrete system representations is given as $z=e^{s\Delta t}$. Since $s$ is a complex variable, one can write $s=\alpha+j\beta$, which implies that $z=e^{\alpha\Delta t}e^{j\beta\Delta t}=e^{\alpha\Delta t}(cos(\beta \Delta t)+jsin(\beta \Delta t))$. Thus, for real values of $s$, the magnitude of $z$ is $|z|=e^{\alpha\Delta t}$, and implies that $e^{\alpha\Delta t}>1$ for $\alpha>0$, $e^{\alpha\Delta t}=1$ for $\alpha=0$ and $e^{\alpha\Delta t}<1$ for $\alpha<0$
        
        However, we do not know what is the expression for $\Phi^k$ (in the same way for continuous system we have that $x(t)=e^{At}x(0)=\left[?\right]x(0)$). Therefore, we can find the state transformation $z(k)=Tx(k)$, so that $D=T\Phi T^{-1}$ is such that
        above Equation becomes:
        $$z(k)=\left[T \Phi T^{-1}\right]z(0) \;\;\;\; \rightarrow z(k)=D^kz(0)=\left[\begin{matrix}\lambda^k_1 & 0 & \cdots& 0\\0 & \lambda^k_2 & 0 & \cdots\\\cdots & \cdots& \cdots &\cdots\\0 & 0& 0& \lambda^k_n \end{matrix}\right]z(0)$$
        
        In order to determine the system's stability (if the state $x(k)$ diverges  or converges for $x_0\neq 0$, as $k\rightarrow \infty$), we have the following criteria: *If* $\Phi$ *is diagonalizable, the necessary condition for above Eq. stability is that all eigenvalues of* $\Phi$, $\|\left\{\lambda_i\right\}\|<1$.
        
        The eigenvalues of $\Phi$ are calculated numerically from  the characteristic polynomial of $\Phi$, that is $\Pi(\lambda)=det(A-\lambda I))=0$,
        $$ \Pi(\lambda)=\lambda^n-p_1\lambda^{n-1}+\cdots+p_{n-1}\lambda +p_n$$
        
        The above Eq. is stable if and only if $\|\lambda\|<1$ for every characteristic root, (that is, for every zero of the characteristic polynomial, see [http://www.youtube.com/watch?v=i9c8MG2KW-g](http://www.youtube.com/watch?v=i9c8MG2KW-g).

#### Example: Determining stability



Determine stability of $x(k+1)=Ax(k)$, where
    $$A=\left[\begin{matrix} 1 & 1\\0.25 &1\end{matrix}\right]$$.

Easily we find
$$P=\left[\begin{matrix} 2 & 2\\1 &-1\end{matrix}\right]$$
so that
$$P^{-1}AP=\left[\begin{matrix} 1.5 & 0\\0 &0.5\end{matrix}\right]$$
thus the system is unstable, see the example in  [http://www.youtube.com/watch?v=i9c8MG2KW-g> ](http://www.youtube.com/watch?v=i9c8MG2KW-g> )how to calculate matrix $P$.

-   Algebraic criteria relate to the properties of the characteristic polynomial (Yuri criteria). Given the characteristic equation in the following form:
    $$\Pi(\lambda)=a_n\lambda^n+a_{n-1}\lambda^{n-1}+\cdots+a_{1}\lambda+a_0$$

The Yuri stability test requires that the following conditions are satisfied in order to guarantee stability:

-   $\Pi(1)>0$
-   $(-1)^n\Pi(-1)>0$ for $n$ even
-   $(-1)^n\Pi(-1)<0$ for $n$ odd
-   and $|a_0|<a_n$, $|b_0|>|b_{n-1}|$, $|c_0|>|c_{n-2}|$,&#x2026;,$|s_0|>|s_{2}|$

  A quick method to determine the stability of the second order system, $\Pi(\lambda)=a_2\lambda^2+a_1 \lambda+a_0$,
is that the following holds:

-   $\Pi(1)>0$
-   $\Pi(-1)>0$
-   $|a_0|<a_2$.
    **Remark**: The alternative technique to the Yury's test that deals with the stability of the characteristic equation is a *modified Routh stability test* which is the stability test to analyze a characteristic equation of the continuous time system. Given the characteristic equation $\Pi(z)=a_nz^n+a_{n-1}z^{n-1}+\cdots+a_{1}z+a_0$ from $\Pi(z)=det(zI-\Phi)$, one can substitute for $z=\frac{1+w}{1-w}$ and obtain the corresponding $\Pi(w)$ which then can be used by standard Routh array test which requires that for stability all coefficients in the Routh array must be positive.

In $Z$-domain, a relaxed LTI system is BIBO stable if and only if its poles of the transfer function $H(z)$ are inside the unit circle. This is a very efficient approach to check the stability of an LTI system. The result is illustrated in the following graph.

![img](./FiguresCHE573/stability_radius2.png "Stability radius")

In CHE 446 Process Control, we learned that for continuous-time systems, if the poles of the Laplace transfer function of a system are on the left half plane of the complex $s$-plane, the system is stable. The difference between the results for continuous-time system and discrete-time systems actually can be explained by studying the relation between Laplace transform and $z$-transform.

Let us recall the definition for Laplace transform. The Laplace transform of a continuous-time signal $x_c(t)$ is defined as follows:

$$X_c(s) = \int_{t=0}^{t=\infty}x_c(t)e^{-st}dt$$

The $z$-transform of a discrete-time signal $x(n)$ is defined as follows:

$$X(z) = \sum_{n=0}^{\infty}x(n)z^{-n}$$

Let us also consider that $x(n)$ is the sampled signal of $x_c(t)$ with the sampling time $T$. Then,

$$x(n)=x_c(nT)$$

Substituting this discretization time back to, we can have:

$$X(z) = \sum_{n=0}^{\infty}x_c(nT)z^{-n}$$

By comparing above equations, it can be seen that:

$$X(z) = X_c(s)|_{t=nT}$$

with

$$z=e^{sT}$$

we can interpret $z$-transform as the sampled Laplace transform with $z=e^{sT}$.

Let us consider an area on the left half of the $s$-plane as shown in the following graph. The area is with the boundary $A\rightarrow B\rightarrow C\rightarrow C'\rightarrow D\rightarrow E'\rightarrow E\rightarrow F\rightarrow A$. In the graph, $w_s=\dfrac{2\pi}{T}$ where $T$ is the sampling time. The points on the boundary are mapped to the $z$-plane as follows:

-   Point A: $s=0$ $\rightarrow$ $z=e^{T\cdot 0}=1$
-   Point B: $s=i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=e^{T\cdot s}=\cos(\dfrac{Tw_s}{2})+i\sin(\dfrac{Tw_s}{2})=\cos(\pi)+i\sin(\pi)=-1$
-   Point C: $s=-\alpha+i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=-e^{-T\alpha}$
-   Point C': $s=-\alpha+i\cdot \dfrac{w_s}{4}$ $\rightarrow$ $z=e^{-T\alpha}(\cos(\dfrac{\pi}{2})+i\sin(\dfrac{\pi}{2}))=i\cdot e^{-T\alpha}$
-   Point D: $s=-\alpha$ $\rightarrow$ $z=e^{-T\alpha}$
-   Point E': $s=-\alpha-i\cdot \dfrac{w_s}{4}$ $\rightarrow$ $z=-i\cdot e^{-T\alpha}$
-   Point E: $s=-\alpha-i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=-e^{-T\alpha}$
-   Point F: $s=-i\cdot \dfrac{w_s}{2}$ $\rightarrow$ $z=-1$

It can be seen that the area in the $s$-plane is mapped into an annulus with outside radius $r_o=1$ and inside radius $r_i=e^{-T\alpha}$. Similar analysis can be applied to other areas. What we should find is that the left half of the $s$-plane is mapped to the inside of the unit circle. This explains why in the discrete-time domain, the stability condition is poles are within the unit circle.
